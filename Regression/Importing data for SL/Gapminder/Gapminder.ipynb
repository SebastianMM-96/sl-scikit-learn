```python
# Import numpy and pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame: df
df = pd.read_csv('gapminder_all.csv')
```

```python
# Create arrays for features and target variable
y = df['life'].values
X = df['fertility'].values
```

```python
# Print the dimensions of X and y before reshaping
print("Dimensions of y before reshaping: {}".format(y.shape))
print("Dimensions of X before reshaping: {}".format(X.shape))
```

```python
# Reshape X and y
y = y.reshape(-1,1)
X_fertility = X.reshape(-1,1)
```

```python
# Print the dimensions of X and y after reshaping
print("Dimensions of y after reshaping: {}".format(y.shape))
print("Dimensions of X after reshaping: {}".format(X_fertility.shape))
```

```python
plt.scatter(X_fertility, y)
plt.ylabel('Life Expectancy')
plt.xlabel('Fertility')
plt.show();
```

```python
# Fitting a regression model
import numpy as np
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
# Fit the model
reg.fit(X_fertility, y)
prediction_space = np.linspace(min(X_fertility), max(X_fertility)).reshape(-1, 1)
```

```python
plt.scatter(X_fertility, y, color = 'blue')
plt.plot(prediction_space, reg.predict(prediction_space), 
color = 'black', linewidth = 3)
plt.show()
```

```python
# Train test split regression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_fertility, y, test_size = 0.3, random_state=42)
```

```python
# Create the regressor: reg_all
reg_all = LinearRegression()

# Fit the regressor to the training data
reg_all.fit(X_train, y_train)

# Predict on the test data: y_pred
y_pred = reg_all.predict(X_test)
```

```python
# Compute and print R^2 and RMSE
print("R^2: {}".format(reg_all.score(X_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: {}".format(rmse))
```
